Features

fossil_size: Size of the fossil, in centimeters.

fossil_weight: Weight of the fossil, in grams.

geological_period: Geological period during which the fossil was formed.

surrounding_rock_type: Type of rock surrounding the fossil.

paleomagnetic_data: Paleomagnetic orientation data of the fossil site.

stratigraphic_position: Position of the fossil within the stratigraphic column.

age: Calculated age of the fossil based on various features, in years.

# Encoding the Categorical columns

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# importing neccessary libraries to convert categoric variables to numeric
# transforming the categoric column into numeric to build machine learning model
```

```python
a = pd.DataFrame() # using the label encoder converting cat to numeric
for i in df_cat.columns:
    a[i]=le.fit_transform(df_cat[i])
```

```python
df_final=pd.concat([df_num,a],axis=1) # combaining all the numeric and encoded columns into one final data Frame.
```

# Base Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # importing library to do linear Regression.
```

```python
from sklearn.model_selection import train_test_split # importing some neccessary syntax to split the dataset and check the 
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_percentage_error # performance of the model.
```

```python
x = df_final.drop(columns='age') # dividing the columns as independent and dependent variables
y=df_final['age']

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3) # splitting the dataset into training and testing.


model = lr.fit(xtrain,ytrain) # fitting the data to the model
ypred=model.predict(xtest)  # predicting the dependent variable with the test data.
print('MSE: ',mean_squared_error(ytest,ypred))# printing the MSE performence metric
print('RMSE: ',np.sqrt(mean_squared_error(ytest,ypred)))  #printing the RMSE performence metric
print('R2: ',r2_score(ytest,ypred)*100)     # printing the r-squared evaluation metric
print('MAPE: ',mean_absolute_percentage_error(ytest,ypred)*100) # printing the MAPE performance metric.
# by seeing the R2,MAPE we can say that the model is performing well. but r2 we have as 97% so it may be overfit.
# lets check by performing ridge regression.
```

```python
# as we got 97% as r2 value we have to check for the adj r2 value which is appropriate for the model.
```

```python
adj_r2=1 - ( 1-r2_score(ytest,ypred) ) * ( len(y) - 1 ) / ( len(y) - x.shape[1] - 1 )
adj_r2*100
```

```python
# By looking at the R2 and Adj_r2 we can say that the model is overfitting. Let's see how it performs when we scale the values..
```

# Linear Regression with Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
# bringing all the values to a signle scale to see how it performs.
```

```python
x = df_final.drop(columns='age')
y=df_final['age'] # building the model with same scale values.and checking the metrics

x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3)


model = lr.fit(xtrain,ytrain)
ypred=model.predict(xtest)
print('MSE: ',mean_squared_error(ytest,ypred))
print('RMSE: ',np.sqrt(mean_squared_error(ytest,ypred))) 
print('R2: ',r2_score(ytest,ypred)*100)
print('MAPE: ',mean_absolute_percentage_error(ytest,ypred)*100)

adj_r2=1 - ( 1-r2_score(ytest,ypred) ) * ( len(y) - 1 ) / ( len(y) - x.shape[1] - 1 )
print('adj_r2',adj_r2)
```

```python
# after scaling also we are getting 97% of r2 and adj_r2 so lets try outlier treatment.
```

# Linear Regression with outlier Treatment

```python
# lets check the performance by removing the outliers.
def outlier_treatmet(df):
    for i in df.columns:
        Q1=df[i].quantile(0.25)
        Q3=df[i].quantile(0.75)
        IQR=Q3-Q1
        UL=Q3 + 1.5*IQR
        LL=Q1-1.5*IQR
        df[i]=np.where(df[i].values>UL,UL,df[i])
        df[i]=np.where(df[i].values<LL,LL,df[i])
    return df
```

```python
x=df_final.drop(columns='age')
x=outlier_treatmet(x)
y=df_final['age']

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3)


model = lr.fit(xtrain,ytrain)
ypred=model.predict(xtest)
print('MSE: ',mean_squared_error(ytest,ypred))
print('RMSE: ',np.sqrt(mean_squared_error(ytest,ypred))) 
print('R2: ',r2_score(ytest,ypred)*100)
print('MAPE: ',mean_absolute_percentage_error(ytest,ypred)*100)

adj_r2=1 - ( 1-r2_score(ytest,ypred) ) * ( len(y) - 1 ) / ( len(y) - x.shape[1] - 1 )
print('adj_r2',adj_r2)
```

```python
# By looking at the r2 and adj r2 , we can say that it is still overfitting the data so lets combine both.
```
